#### softmax

```python
import numpy as np
```

```python
a = np.array([0.3,2.9,4.0])
exp_a = np.exp(a) # 指数函数
print(exp_a)
```

```python
sum_exp_a = np.sum(exp_a)
print(sum_exp_a)
```

```python
y = exp_a / sum_exp_a
print(y)
```

```python
def softmax(a):
    exp_a = np.exp(a)
    sum_exp_a = np.sum(exp_a)
    y = exp_a / sum_exp_a

    return y
```

```python
a = np.array([1010,1000,990])
print(np.exp(a) / np.sum(np.exp(a)))
```

```python
c = np.max(a)
print(a-c)
print(np.exp(a-c) / np.sum(np.exp(a-c)))
```

```python
def softmax(a):
    c = np.max(a)
    exp_a = np.exp(a-c)
    sum_exp_a = np.sum(exp_a)
    y = exp_a / sum_exp_a
    return y
```

```python
a = np.array([0.3,2.9,4.0])
y = softmax(a)
print(y)
```
