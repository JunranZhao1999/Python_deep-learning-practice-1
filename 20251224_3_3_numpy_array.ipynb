#### 数组乘法

```python
A = np.array([[1,2],[3,4]])
B = np.array([[5,6],[7,8]])
A * B
```

```python
# 点乘
np.dot(A,B)
```

#### 2 × 3 的矩阵和 3 × 2 的矩阵的乘积可按如下形式用 Python 来实现。

```python
A = np.array([[1,2,3],[4,5,6]])
A.shape
```

```python
A.ndim
```

```python
B = np.array([[1,2],[3,4],[5,6]])
B.shape
```

```python
B.ndim
```

```python
np.dot(A,B)
```

```python
np.dot(B,A)
```
