#### 阶跃函数的图形

```python
import numpy as np
import matplotlib.pyplot as plt

def step_function(x):
    return np.array(x > 0, dtype = int)

x = np.arange(-5.0,5.0,0.1)
y = step_function(x)

plt.plot(x,y)
plt.ylim(-0.1,1.1) # 指定 y 轴的范围
plt.show()
```

#### sigmond 函数

```python
def sigmond(x):
    return 1 / (1 + np.exp(-x))
```

```python
x = np.array([-1.0,1.0,2.0])
sigmond(x)
```

```python
t = np.array([1.0,2.0,3.0])
print(1.0 + t)
print(1.0 / t)
```

```python
x = np.arange(-5.0,5.0,0.1)
y = sigmond(x)

plt.plot(x,y)
plt.ylim(-0.1,1.1)
plt.show()
```

```python
import seaborn as sns

sns.set_palette('crest')

x = np.arange(-5.0,5.0,0.1)
y = step_function(x)
plt.plot(x,y,linestyle = ':')

x = np.arange(-5.0,5.0,0.1)
y = sigmond(x)
plt.plot(x,y)

plt.ylim(-0.1,1.1)
plt.show()
```

#### relu

```python
def relu(x):
    return np.maximum(0,x)
```

```python
x = np.arange(-5.0,5.0,0.1)
y = relu(x)

plt.plot(x,y)
plt.ylim(-0.1,5.1)
plt.show()
```
