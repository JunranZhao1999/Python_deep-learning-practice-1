### 访问元素

元素的索引从 0 开始。对各个元素的访问可按如下方式进行。

```python
X = np.array([[51,55],[14,19],[0,4]])
X
```

```python
X[0] # 第0行
```

```python
X[0][1] # (0,1)的元素
```

也可以使用 for 语句访问各个元素。

```python
for row in X:
    print(row)
```

除了前面介绍的索引操作,NumPy 还可以使用数组访问各个元素。

```python
X = X.flatten()
X
```

```python
X[np.array([0,2,4])] # 获取索引为 0、2、4 的元素
```
